```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import os
import PIL
```

```python
path = 'traffic signs'

count = 0
pixels = []
labels = []
my_list = os.listdir(path)
print("Total classes detected: ", len(my_list))
no_of_classes = len(my_list)
print("Importing classes...")

for x in range(0, len(my_list)):
    my_pic_lst = os.listdir(path+"/"+str(x))
    for y in my_pic_lst:
        cur_img = PIL.Image.open(path+'\\'+str(x)+'\\'+y)
        image = cur_img.resize((32,32))
        image = np.array(image)
        images.append(image)
        class_no.append(count)
    print(count, end=' ')
    count += 1
    
pixels = np.array(images)
labels = np.array(class_no)

data = {}

data['pixels'] = pixels
data['labels'] = labels

import pickle 
pickle.dump(data, open('data.p', 'ab'))
```
